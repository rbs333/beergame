```python
import pandas as pd
```

```python
ret = pd.read_csv("PlayerType.RETAILER_log.csv")
sup = pd.read_csv("PlayerType.SUPPLIER_log.csv")
dist = pd.read_csv("PlayerType.DISTRIBUTOR_log.csv")
manf = pd.read_csv("PlayerType.MANUFACTURER_log.csv")

res = [ret, sup, dist, manf]

ret['player'] = "retailer"
sup['player'] = "supplier"
dist['player'] = "distributor"
manf['player'] = "manufacturer"
```

```python
all_players = ret.append(sup)
```

```python
all_players = pd.DataFrame(columns=ret.columns[1:])

for r in res:
    all_players = all_players.append(r, ignore_index=True)

all_players = all_players[all_players["round"] != 16]
```

```python
cost_view = all_players.pivot(index="round", columns="player", values="round_cost")
cost_view.plot(figsize=(10,5))
```

```python
demand_view = all_players.pivot(index="round", columns="player", values="demand")
demand_view.plot(figsize=(10,5))
```

```python
demand_view = all_players.pivot(index="round", columns="player", values="end_inv")
demand_view.plot(figsize=(10,5))
```
